# **Despliegue: Car Risk**
- Cargamos el modelo
- Cargamos los datos futuros
- Preparar los datos futuros
- Aplicamos el modelo de predicción

```python
#Cargamos librerías principales

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pickle
filename = 'modelo-clas-tree-knn-nn.pkl'
modelTree,modelKnn,modelNN,labelencoder,variables,min_max_scaler = pickle.load(open(filename, 'rb')) #read
```

```python
#Cargamos los datos futuros

data = pd.read_csv("carRisk-future.csv", sep = ",")
data.head()
```

```python
#Se realiza la preparación creando dummis

data_preparada=data.copy()
data_preparada = pd.get_dummies(data_preparada, columns=['cartype'], drop_first=False)
data_preparada.head()
```

```python
#Se adicionan las columnas faltantes en este caso Minivan

data_preparada=data_preparada.reindex(columns=variables,fill_value=0)

data_preparada.head()
```

# **Predicción con Tree**

```python
#Hacemos la predicción con Tree

Y_fut = modelTree.predict(data_preparada)

print(Y_fut)
```

```python
print (labelencoder.inverse_transform(Y_fut))
```

```python
data['Prediccion Tree']=labelencoder.inverse_transform(Y_fut)

data
```

# **Prediccion con KNN**

```python
#Normalización

data_preparada[['age']]= min_max_scaler.transform(data[['age']]) 
data_preparada.head()
```

```python
#Hacemos la predicción con knn

Y_fut = modelKnn.predict(data_preparada)

data['Prediccion Knn']=labelencoder.inverse_transform(Y_fut)

data
```

# **Neuural Network**

```python
#Hacemos la predicción con la NN
Y_fut = modelNN.predict(data_preparada)
data['Prediccion NN']=labelencoder.inverse_transform(Y_fut)
data
```
